#Import libraries and load a pre-trained model:

```python
import nltk
from nltk.tokenize import word_tokenize
from transformers import pipeline
```

```python
nltk.download("punkt")
```

```python
nlp = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")
```

#Define a function for question answering:

```python
def answer_question(question, context):
    result = nlp(question=question, context=context)
    return result["answer"]
```

#Use the function to answer questions from your dataset:

```python
dataset = [
    ("What is the role of insulin in diabetes?", "Insulin's role in the pathogenesis of diabetes mellitus."),
    ("How does aspirin affect cardiovascular health?", "Effects of aspirin on cardiovascular health."),
    ("What are the symptoms of COVID-19?", "Common symptoms of COVID-19 include fever, cough, and difficulty breathing."),
    ("What are the risk factors for lung cancer?", "Common risk factors for lung cancer include smoking, exposure to radon gas, and occupational exposures to carcinogens."),
    ("How is HIV transmitted?", "HIV is primarily transmitted through unprotected sexual intercourse, sharing needles for drug use, and from mother to child during childbirth or breastfeeding."),
    ("What is the mechanism of action of antibiotics?", "Antibiotics work by targeting and inhibiting specific components of bacterial cells, such as cell walls or protein synthesis machinery."),
]
```

```python
for question, context in dataset:
    answer = answer_question(question, context)
    print(f"Question: {question}")
    print(f"Answer: {answer}\n")
```
